```python
pengZhang=0.03#假设通胀率固定
deadYear=80#假设死亡年龄
avMonthSpend=0.4#假设退休后 每月花费
avMonthEarned=0.3#假设退休后 每月收入
deadWith=20#假设死亡时剩余的保底金额or活得比deadYear长，继续存活的保底金额 单位：w
interest=0.035#假设存款利率固定
currentAge=25#假设开始工作&&现在的年龄

#-----------------其他前提-------------------
#1.有房且不用再买，不考虑养车费or买车费
#2.avMonthEarned为退休金和简单工作工资的总和
#3.旅游费用+养房费用+大病花费，包含在每月花费avMonthSpend中
#4.保险已经购买，且保险均不为理财型保险

#-----------------二选一参与计算-------------------
retireYear=35#假设此时选择简单的工作
beginSaving=100#假设退休时的初始存款 单位：w


#想要计算得到：
#retireYear=？ #假设此时选择简单的工作
#beginSaving=？ #假设退休时的初始存款 单位：w
```

```python
yearOrSaving=True#达到年龄或达到存款额，开始转换工作
workYear=25#开始工作的年龄
avYearSaving=12#假设工作时 每年存的钱 单位：w

#想要计算得到：
#avYearSaving=10#假设工作时 每年存的钱 单位：w
```

```python
#---------------------必要前提！钱越花越少，同一年的saving12必然小于saving1 ！！！-----------------------

if yearOrSaving==True:#假设：retireYear岁时，开始转换工作，半退休
    saving12=deadWith*((1+pengZhang)**(deadYear-currentAge))
    print(deadYear,'岁保底金额：',deadWith,'万元，通胀后为：',round(saving12,2),'万元')
    for i in range(0,deadYear-retireYear):
        realAge=deadYear-i
        realRate=(1+pengZhang)**(realAge-currentAge)
        cost=(avMonthSpend-avMonthEarned)*12*realRate
        
        #按最差的情况计算，每年1月取上一年的利息&预留一年所花的钱（上一年的存款以saving12 12月份存款为计）
        #saving1=saving12+cost
        income=saving12*interest
        real=cost-income
        saving1=saving12+real
        print(realAge,'岁那一年\t从',round(saving1,2),'花到',round(saving12,2),
              '\t\t年花销',round(cost*10000),' \t通胀了',round(realRate,2),'倍\t实际花费：',
              round(cost*10000),'-',round(income*10000),'=',round(real*10000))
        saving12=saving1
    realSaving=saving1/((1+pengZhang)**(retireYear-currentAge))
    print('以上是',retireYear,'岁退休的生活计划，',retireYear,'岁时的',round(saving1,2),'w相当于现在的',round(realSaving,2),'w')    

        
```

```python
deposit=saving1
#假设每年年底存入avYearSaving元，从workYear到retireYear
for i in range(1,retireYear-workYear):
    if(deposit<0): break
    toBE=deposit
    realAge=retireYear-i
    deposit=deposit/(1+interest)
    #avYearSaving 每年不涨薪
    print(realAge,'岁那一年年末需要存入',round(deposit,2),'万元（含那年净收入',avYearSaving,' 万元）\t3.5%利率膨胀到',round(toBE,2))
    deposit=deposit-avYearSaving
```
